```python
#@markdown Sounds (TTS)
!pip install gtts


from gtts import gTTS
from IPython.display import Audio

def etts(mytext):
    tts = gTTS(mytext, lang="en", tld = "us")
    tts.save('sample.mp3')
    return Audio('sample.mp3', autoplay=True)
```

```python
types = "1. -ity" #@param = ["1. -ity", "2. -y", "3. -ive", "4. -(i)al", "5. -ize", "6. ion"]
t = types.split(".")
t1 = t[0]
t2 = int(t1)-1


wlist = ["1. public versus publicity. 2. sane versus sanity. 3. Clear versus clarity. 4. flexible versus flexibility. 5. Humble versus humility.",
         "1. democrat versus democracy. 2. photo versus photography. 3. analog versus analogy. 4. Prophet versus prophecy. 5. allergic versus allergy. ",
         "1. product versus productive. 2. object versus objective. 3. project versus projective. 4. select versus selective. retract versus retractive.",
         "1. part vs. partial. 2. essens vs. essential. 3. torrent vs. torrential.",
         "1. critic vs. criticise. 2. romantic vs. romanticize. 3. plastic vs. plasticize. 4. logic vs. logicize. ",
         "1. native vs. nation. 2. relative vs. relation. 3. creative vs. creation. 4. innovative vs. innovation. 5. repetitive vs. repetition."]

mytext = wlist[t2]

etts(mytext)
Audio('sample.mp3', autoplay = True)
```
