## [1] Creating audio sample on colab

To create an audio sample using Python code in Google Colab, you will need to use the IPython.display module and the Audio function. Here is an example of how you can create a simple audio sample using a sine wave:

```python
!pip IPython.display
```

```python
from IPython.display import Audio
import numpy as np

# Generate a sine wave at 440 Hz
frequency = 440
sample_rate = 44100
duration = 3  # 3 seconds

t = np.linspace(0, duration, sample_rate * duration)
waveform = np.sin(2 * np.pi * frequency * t)

# Create the audio sample
audio = Audio(waveform, rate=sample_rate)
display(audio)
```

This code will create a 3-second audio sample of a sine wave at 440 Hz. The rate parameter specifies the sample rate, which determines the quality and resolution of the audio.

## [2] Creating a vowel-like sound

```python
from IPython.display import Audio
import numpy as np

# Generate a sine wave at the frequency of the vowel sound in "mom"
frequency = 330  # Hz
sample_rate = 44100  # samples per second
duration = 0.5  # 0.5 seconds

t = np.linspace(0, duration, sample_rate * duration)

waveform = np.sin(2 * np.pi * frequency * t)

# Create the audio sample
audio = Audio(waveform, rate=sample_rate)
display(audio)
```

```python
from IPython.display import display, Image
from PIL import Image, ImageDraw
import time

# Create a series of images
images = []
for i in range(10):
  img = Image.new('RGB', (200, 200), color='white')
  draw = ImageDraw.Draw(img)
  draw.text((20, 20), f'Frame {i}', fill='black')
  images.append(img)

# Display the images as an animation
for img in images:
  display(img)
  time.sleep(0.5)
```
